```python
import requests
import pandas as pd

# Stock symbols as a string, separated by commas
stocks = input("Enter stock symbols separated by commas (e.g., AAPL,GOOGL,AMZN): ")   # Add more stocks separated by commas

# API Key for Yahoo Finance API (replace with your actual API key)
apikey = ""  # Replace with your actual API key

# Define the URL for the Yahoo Finance API
url = "https://yfapi.net/v6/finance/quote"
url2 = "https://yfapi.net/v6/finance/quote"

# Query parameters with the stock symbols
querystring = {"symbols": stocks}

# Set up the headers with the API key
headers = {
    'x-api-key': apikey
}


response = requests.get(url, headers=headers, params=querystring)
stock_data = response.json().get('quoteResponse', {}).get('result', [])
df = pd.DataFrame(stock_data)
print("Basic Stock Data:\n")
print(df[['symbol', 'longName', 'regularMarketPrice']])

# TASK 1
for stock in stock_data:
    stock_ticker = stock.get('symbol', 'N/A')
    company_name = stock.get('longName', 'N/A')
    current_price = stock.get('regularMarketPrice', 'N/A')
    print(f"Stock Ticker: {stock_ticker}")
    print(f"Company: {company_name}")
    print(f"Current Market Price: ${current_price:,.2f}")
    print("-" * 50)

# TASK 2
module = int(input('Input Module (1 for 52 Week Data, 2 for Return on Assets): '))
data_list=[]
if module == 1:
    response = requests.get(url, headers=headers, params=querystring)
    task2data = response.json().get('quoteResponse', {}).get('result', [])
    df = pd.DataFrame(task2data)

    for task2 in task2data:
        stock_ticker2 = task2.get('symbol', 'N/A')
        company_name2 = task2.get('longName', 'N/A')
        fiftyTwoWeekH = task2.get('fiftyTwoWeekHigh', 'N/A')
        fiftyTwoWeekL = task2.get('fiftyTwoWeekLow', 'N/A')

        data_list.append({
            "Stock Ticker": stock_ticker2,
            "Company": company_name2,
            "52 Week High": fiftyTwoWeekH,
            "52 Week Low": fiftyTwoWeekL
        })

elif module == 2:
    response = requests.get(url2, headers=headers, params=querystring)
    task2data = response.json().get('quoteResponse', {}).get('result', [])
    df = pd.DataFrame(task2data)

    for task2 in task2data:
        stock_ticker2 = task2.get('symbol', 'N/A')
        company_name2 = task2.get('longName', 'N/A')
        #Return on Assets doesn't exist in any yahoo finance api, so the next closest thing is regular market change percent
        regularMarketChangePercent= task2.get('regularMarketChangePercent', 'N/A')

        data_list.append({
            "Stock Ticker": stock_ticker2,
            "Company": company_name2,
            "Return on Assets (approx)": regularMarketChangePercent
        })
df = pd.DataFrame(data_list)
print("\nDataFrame Result:\n", df)

#TASK 3

region = input("Input Region, must be 'us' or 'hk' ")
url3 = f'https://yfapi.net/v1/finance/trending/{region}'


response = requests.get(url3, headers=headers)

response_data = response.json()


task3data = response_data.get('finance', {}).get('result', [])[0].get('quotes', [])
new_list1=[]
df = pd.DataFrame(task3data)
for stock in task3data:
    stock_ticker3 = stock.get('symbol', 'N/A')
    new_list1.append(stock_ticker3)

trending_ticker1=new_list1[0]
trending_ticker2=new_list1[1]
trending_ticker3=new_list1[2]

trending_tickers= trending_ticker1 + "," + trending_ticker2 + "," +trending_ticker3

querystring = {"symbols": trending_tickers}

response = requests.get(url, headers=headers, params=querystring)
stock_data = response.json().get('quoteResponse', {}).get('result', [])
df = pd.DataFrame(stock_data)
print("Top 3 Trending Tickers in "+ region + ":\n")

for trending_tickers in stock_data:
    stock_ticker = trending_tickers.get('symbol', 'N/A')
    company_name = trending_tickers.get('longName', 'N/A')
    current_price = trending_tickers.get('regularMarketPrice', 'N/A')
    fiftyTwoWeekH = trending_tickers.get('fiftyTwoWeekHigh', 'N/A')
    fiftyTwoWeekL = trending_tickers.get('fiftyTwoWeekLow', 'N/A')
    print(f"Stock Ticker: {stock_ticker}")
    print(f"Company: {company_name}")
    print(f"Current Market Price: ${current_price:,.2f}")
    print(f"52 Week High: ${fiftyTwoWeekH}")
    print(f"52 Week Low: ${fiftyTwoWeekL}")
    print("-" * 50)

```
